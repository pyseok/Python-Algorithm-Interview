### 연결리스트를 뒤집어라

입력\
1->2->3->4->5->Null

출력\
5->4->3->-2>-1->Null

```python
def reverseList(head : Node)->Node:
    def reverse(node : Node, prev : Node=None):
        if not node:
            return prev
        next, node.next = node.next, prev
        return reverse(next, node)
    return reverse(head)
```

### 반복 구조로 뒤집기

```python
def reverseList(head : Node) -> Node:
    node, prev = head, None
    
    while node:
        next, node.next = node.next, prev
        prev, node = node, next
    return prev
```
